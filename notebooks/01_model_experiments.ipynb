# 01 Model Experiments
This notebook loads data, trains SVM and Isolation Forest models, and evaluates fraud detection performance.

```python
import pandas as pd
from fraud_detection.config import FEATURE_COLUMNS, TARGET_COLUMN, SAMPLE_FRAC, TEST_SIZE, RANDOM_STATE
from fraud_detection.data_loader import load_huggingface_dataset
from fraud_detection.features import select_features, sample_and_split, scale_features
from fraud_detection.train import train_svm, train_isolation_forest
from fraud_detection.evaluate import evaluate_classifier, evaluate_anomaly_detector
```

```python
# Option 1: load from Hugging Face
# df = load_huggingface_dataset('electricsheepafrica/Nigerian-Financial-Transactions-and-Fraud-Detection-Dataset')

# Option 2: load a local CSV from data/raw/
df = pd.read_csv('../data/raw/your_dataset.csv')
df.head()
```

```python
X, y = select_features(df, FEATURE_COLUMNS, TARGET_COLUMN)
X_train, X_test, y_train, y_test = sample_and_split(X, y, sample_frac=SAMPLE_FRAC, test_size=TEST_SIZE, random_state=RANDOM_STATE)
scaler, X_train_scaled, X_test_scaled = scale_features(X_train, X_test)
X_train.shape, X_test.shape
```

```python
svm_model = train_svm(X_train_scaled, y_train, random_state=RANDOM_STATE)
svm_results = evaluate_classifier(svm_model, X_test_scaled, y_test)
svm_results['roc_auc']
```

```python
iso_model = train_isolation_forest(X_train_scaled, y_train, contamination=0.01, random_state=RANDOM_STATE)
iso_results = evaluate_anomaly_detector(iso_model, X_test_scaled, y_test)
iso_results['roc_auc']
```

```python
pd.DataFrame({
    'model': ['SVM', 'Isolation Forest'],
    'roc_auc': [svm_results['roc_auc'], iso_results['roc_auc']]
})
```
